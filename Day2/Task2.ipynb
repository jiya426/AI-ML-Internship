## Objective
To analyze the Titanic dataset using statistical summaries and visualizations to understand patterns and relationships.

## Tools Used
- Pandas
- Matplotlib
- Seaborn



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

## Summary Statistics

We generate descriptive statistics to understand central tendency and spread of numerical features.


```python
df.describe()
```

```python
df.mean(numeric_only=True)
```

```python
df.median(numeric_only=True)
```

```python
df.std(numeric_only=True)
```

## Distribution of Numerical Features

Histograms help visualize data distribution and detect skewness.


```python
df.hist(figsize=(10,8))
plt.show()
```

### Observations:
- Age distribution is slightly right-skewed.
- Fare shows extreme right skewness.
- Most passengers traveled alone (SibSp = 0).


## Outlier Detection Using Boxplots


```python
sns.boxplot(x=df['Fare'])
plt.show()
```

### Observation:
Fare contains significant outliers indicating some passengers paid extremely high ticket prices.


## Survival Analysis by Gender


```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.show()
```

### Observation:
Female passengers had a much higher survival rate compared to males.


## Survival Analysis by Passenger Class


```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.show()
```

### Observation:
Passengers in 1st class had a higher survival rate compared to 3rd class.


## Correlation Matrix


```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()
```

### Observation:
Fare has positive correlation with survival.
Passenger class is negatively correlated with survival.


## Pairplot for Feature Relationships


```python
sns.pairplot(df[["Age","Fare","Survived"]], hue="Survived")
plt.show()

```

```python
df.skew(numeric_only=True)
```

## Skewness Analysis

Positive skewness indicates right-tailed distribution.
Fare shows high positive skewness.


## Final Insights from EDA

- Female passengers had higher survival rates.
- Higher class passengers were more likely to survive.
- Fare is strongly right-skewed and contains outliers.
- Passenger class negatively impacts survival probability.
- Most passengers were between 20â€“40 years old.
